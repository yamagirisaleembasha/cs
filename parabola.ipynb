```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
x=list(map(float,input('enter x values:').split()))
y=list(map(float,input('enter y values:').split()))
x=np.array(x)
y=np.array(y)
xy=x*y
x_square=x**2
x_square_y=x_square*y
x_cube=x**3
x_power_four=x**4
mat=np.array([[len(x),sum(x),sum(x_square)],[sum(x),sum(x_square),sum(x_cube)],[sum(x_square),sum(x_cube),sum(x_power_four)]])
mat_1=np.array([[sum(y),sum(xy),sum(x_square_y)],[sum(x),sum(x_square),sum(x_cube)],[sum(x_square),sum(x_cube),sum(x_power_four)]])
mat_2=np.array([[len(x),sum(x),sum(x_square)],[sum(y),sum(xy),sum(x_square_y)],[sum(x_square),sum(x_cube),sum(x_power_four)]])
mat_3=np.array([[len(x),sum(x),sum(x_square)],[sum(x),sum(x_square),sum(x_cube)],[sum(y),sum(xy),sum(x_square_y)]])
det=np.linalg.det(mat)
det_1=np.linalg.det(mat_1)
det_2=np.linalg.det(mat_2)
det_3=np.linalg.det(mat_3)
a=round(det_1/det,2)
b=round(det_2/det,2)
c=round(det_3/det,2)
y_cal=[]
for i in range(len(x)):
    y_cal.append(round(a+(b*x[i])+(c*(x[i]**2)),2))
plt.plot(y,x,marker='o',c='g')
plt.plot(y_cal,x,marker='d',linestyle="--",c='r')
plt.show
```
