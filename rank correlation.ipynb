```python
print('input:')
import pandas as pd
import numpy as np
import scipy.stats as st
import math
x_input=list(map(int,input("enter x values:").split()))
y_input=list(map(int,input("enter y values:").split()))
print()
setx=set(x_input)
sety=set(y_input)
if len(setx)!=len(x_input) or len(sety)!=len(y_input):
    rankx=pd.Series(x_input).rank()
    ranky=pd.Series(y_input).rank()
    A=len(x_input)-rankx+1
    B=len(y_input)-ranky+1
    d={'X':x_input,'Y':y_input,'A':list(A),'B':list(B)}
    a=pd.DataFrame(d)
    A=np.array(a['A'])
    B=np.array(a['B'])
    di=A-B
    disquare=di**2
    disquaresum=sum(disquare)
    a['di']=di
    a['disquare']=disquare
    print('output:')
    print()
    print('given data has repeated observations')
    print()
    print('dataframe is:')
    print(a)
    print()
    dictx={}
    dicty={}
    for i in range(len(x_input)):
        if x_input[i] not in dictx:
            dictx[x_input[i]]=1
        else:
            dictx[x_input[i]]+=1
        if y_input[i] not in dicty:
            dicty[y_input[i]]=1
        else:
            dicty[y_input[i]]+=1
    m=[]
    for keys in dictx.values():
          if keys>1:
                m.append(keys)
    for ykeys in dicty.values():
          if ykeys>1:
                m.append(ykeys)
    cf=0
    for m1 in m:
        cf+=m1*((m1**2)-1)/12
    modified_disquaresum=disquaresum+cf
    rho=1-(6*modified_disquaresum)/(len(x_input)*(len(x_input)**2-1))
    rho=round(rho,4)
    print('rank correlation of the given data is:',rho)
else:
    ranks_of_x=pd.Series(x_input).rank()
    ranks_of_y=pd.Series(y_input).rank()
    n=len(x_input)
    ranks_of_x=n-ranks_of_x+1
    xranks=np.array(ranks_of_x)
    ranks_of_y=n-ranks_of_y+1
    yranks=np.array(ranks_of_y)
    di=xranks-yranks
    disquare=np.array(di**2)
    disquaresum=sum(disquare)
    row=1-((6*disquaresum)/(n*(n**2-1)))
    row=round(row,4)
    x=pd.DataFrame({'X':x_input,'Y':y_input,'A':list(ranks_of_x),'B':list(ranks_of_y),'di=A-B':list(di),'di^2':list(disquare)})
    print('output:')
    print()
    print('given data has repeated observations')
    print()
    print('dataframe is:')
    print(x)
    print()
    print('rank correlation of the given data is:',row)
print()
print('rank correlation of the given data is by built in function is:',round(st.spearmanr(x_input, y_input)[0],4))
```
