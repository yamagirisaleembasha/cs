```python
import pandas as pd
p=pd.read_excel('Z://y20cs190_saleembasha//2 nd year//sem4//CS261 CS LAB//marks.xlsx')
p
```

```python
p['total']=p[211]+p[212]+p[213]+p[214]+p[215]
p
```

```python
p['percentage']=(p['total']/500)*100
p
```

```python
grade=[]
for i in p['percentage']:
    if i>=90:
        grade.append('A+')
    elif i>=80:
        grade.append("A")
    elif i>=70:
        grade.append("B") 
    elif i>=60:
        grade.append("C")
    elif i>=50:
        grade.append("D")
    elif i>=40:
        grade.append("E")
    elif i<40:
        grade.append("F")
p['Grade']=grade
```
